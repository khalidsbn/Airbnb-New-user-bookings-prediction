```python
import pandas as pd
from sklearn.metrics import ndcg_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import label_binarize
from sklearn.ensemble import RandomForestClassifier
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier


import warnings
warnings.filterwarnings('ignore')
```

### Prepare data

```python
train_df = pd.read_csv('../app/static/data/users_cleaned_df_project1.csv')
target = pd.read_csv('../app/static/data/target.csv')

X = train_df
y = target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

### Random forest Algorithm

```python
forest = RandomForestClassifier(class_weight='balanced', random_state=42)
forest.fit(X_train, y_train)

y_proba_train_rf = forest.predict_proba(X_train)
y_proba_test_rf = forest.predict_proba(X_test)

y_train_multilabel_rf = label_binarize(y_train, classes=range(12))
y_test_multilabel_rf = label_binarize(y_test, classes=range(12))

# Score:
rf_train_score = ndcg_score(y_train_multilabel_rf, y_proba_train_rf, k=5)
rf_train_score = f"{rf_train_score * 100:.2f}"
print(f'XGBoost NDCG score (Train): {rf_train_score}%')

rf_test_score = ndcg_score(y_test_multilabel_rf, y_proba_test_rf, k=5)
rf_test_score = f"{rf_test_score * 100:.2f}"
print(f'XGBoost NDCG score (Test): {rf_test_score}%')

pd.Series(forest.feature_importances_, index=X_train.columns)[:20].sort_values().plot(title = 'Feature Importances', kind='barh');
```

### LightGBM Algorithm


```python
lgb_clf = LGBMClassifier(objective='multiclass', random_state=42)
lgb_clf.fit(X_train, y_train)

y_proba_train_lgb = lgb_clf.predict_proba(X_train)
y_proba_test_lgb = lgb_clf.predict_proba(X_test)

y_train_multilabel_lgb = label_binarize(y_train, classes=range(12))
y_test_multilabel_lgb = label_binarize(y_test, classes=range(12))


# Score:
print('\n')
lgb_train_score = ndcg_score(y_train_multilabel_lgb, y_proba_train_lgb, k=5)
lgb_train_score = f"{lgb_train_score * 100:.2f}"
print(f'XGBoost NDCG score (Train): {lgb_train_score}%')

lgb_test_score = ndcg_score(y_test_multilabel_lgb, y_proba_test_lgb, k=5)
lgb_test_score = f"{lgb_test_score * 100:.2f}"
print(f'XGBoost NDCG score (Test): {lgb_test_score}%')

pd.Series(lgb_clf.feature_importances_, index=X_train.columns)[:20].sort_values().plot(kind='barh');
```

### XGBoost Algorithm


```python
xgb = XGBClassifier(n_estimators=5, random_state=42)
xgb.fit(X_train, y_train)

y_proba_train_xgb = xgb.predict_proba(X_train)
y_proba_test_xgb = xgb.predict_proba(X_test)

y_train_multilabel_xgb = label_binarize(y_train, classes=range(12))
y_test_multilabel_xgb = label_binarize(y_test, classes=range(12))

# Score:
xgb_train_score = ndcg_score(y_train_multilabel_xgb, y_proba_train_xgb, k=5)
xgb_train_score = f"{xgb_train_score * 100:.2f}"
print(f'XGBoost NDCG score: {xgb_train_score}%')

xgb_test_score = ndcg_score(y_test_multilabel_xgb, y_proba_test_xgb, k=5)
xgb_test_score = f"{xgb_test_score * 100:.2f}"
print(f'XGBoost NDCG score: {xgb_test_score}%')

pd.Series(xgb.feature_importances_, index=X_train.columns)[:20].sort_values().plot(title='Feature Importances', kind='barh');
```

### Compare Results

```python
print('\tAlgorithm\t |\t Train Score \t | \t Test Score')
print('\t----------------------------------------------------------------')
print(f'\tRandom Forest  \t |\t {rf_train_score}% \t | \t {rf_test_score}%')
print(f'\tLightGBM  \t |\t {lgb_train_score}% \t | \t {lgb_test_score}%')
print(f'\tXGBoost  \t |\t {xgb_train_score}% \t | \t {xgb_test_score}%')
```
